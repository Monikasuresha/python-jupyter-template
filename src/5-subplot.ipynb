### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

# Create a figure with 2x2 subplots
fig, axs = plt.subplots(2, 2)

# First subplot
axs[0, 0].plot(x, y1, 'tab:blue')
axs[0, 0].set_title('Line Plot')

# Second subplot
axs[0, 1].bar(x, y2, color='tab:orange')
axs[0, 1].set_title('Bar Chart')

# Third subplot
axs[1, 0].scatter(x, y1, color='tab:green')
axs[1, 0].set_title('Scatter Plot')

# Fourth subplot
axs[1, 1].pie([15, 30, 45, 10], labels=['A', 'B', 'C', 'D'], autopct='%1.1f%%')
axs[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()

# Display the plots
plt.show()
```
